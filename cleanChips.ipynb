Cleaning the cropped 'chips' to prepare for feature extraction

```python
from imports import *
```

```python
root = 'cropped_images/'

chip_paths = os.listdir('cropped_images')
chip_paths = [root + i for i in chip_paths]
```

```python
img_path = chip_paths[555]
img = cv2.imread(img_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

img_nobg = remove(img_rgb)

fig, axs = plt.subplots(1, 2, figsize=(10, 5))
axs[0].imshow(img_rgb)
axs[0].axis('off')
axs[1].imshow(img_nobg)
axs[1].axis('off')
plt.show()
```

```python
sift = cv2.SIFT_create()
keypoints, descriptors = sift.detectAndCompute(img_nobg, None)

img_sift = cv2.drawKeypoints(img_rgb, keypoints, None, flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

plt.figure(figsize=(10, 5))
plt.imshow(img_sift)
plt.axis('off')
plt.show()
```
