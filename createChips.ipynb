```python
root_dir = "2013_tiger_kanha"
li = os.listdir(root_dir)

images = []
for i in tqdm(li):
    folder_img = os.listdir(root_dir + '/' + i)
    for j in folder_img:
        images.append(cv2.imread(root_dir + '/' + i + '/' + j))
```

```python
model = YOLO('models/yolov8n.pt')
```

```python
if os.path.exists("cropped_images"):
    for file in os.listdir("cropped_images"):
        os.remove(os.path.join("cropped_images", file))
else:
    os.makedirs("cropped_images")
```

```python
for i in tqdm(range(len(images))):
    results = model(images[i])[0]
    
    boxes = results.boxes.xyxy.cpu().numpy()
    for j in boxes:
        x1, y1, x2, y2 = map(int, j)

        if (x2 - x1) < 100 or (y2 - y1) < 100:
            continue
        cropped_img = images[i][y1:y2, x1:x2]
        cv2.imwrite(f"cropped_images/cropped_{i}_{j}.jpg", cropped_img)

        # clear_output(wait=True)

        # plt.figure(figsize=(10, 5))
        # plt.subplot(1, 2, 1)
        # plt.imshow(cv2.cvtColor(images[i], cv2.COLOR_BGR2RGB))
        # plt.axis('off')

        # plt.subplot(1, 2, 2)
        # plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
        # plt.axis('off')

        # plt.tight_layout()

        # plt.show()
        # input()
```
